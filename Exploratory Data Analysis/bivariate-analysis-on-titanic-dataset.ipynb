```python
import pandas as pd
import seaborn as sns
```

```python
df =pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
df.head()
```

# Countplot

```python
sns.countplot(df['Embarked'])
```

# Piechart

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
```

# Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

# Displot

```python
sns.distplot(df['Age'])
```

# Boxplot

```python
sns.boxplot(df['Age'])
```

# Bivariate/Multivariate Analysis

# Scatterplot

```python
sns.scatterplot(df['Age'],df['Pclass'], hue=df['Sex'],style=df['Survived'])
```

# Boxplot

```python
sns.boxplot(x=df['Sex'],y=df['Age'],hue=df['Survived'])
```

# Displot

```python
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
```

# Heatmap

```python
pd.crosstab(df['Pclass'],df['Survived'])
```

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

# Clustermap

```python
sns.clustermap(pd.crosstab(df['Parch'],df['Survived']))
```

```python
pd.crosstab(df['Parch'],df['Survived'])
```
